```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#get the data
data = pd.read_csv("../Daily Bike Sharing.csv", 
                   index_col = "dteday", 
                   parse_dates = True)
data.head(1)
```

```python
#select variables
dataset = data.loc[:, ["cnt", "holiday", "workingday", "weathersit",
                       "temp", "atemp", "hum", "windspeed"]]
dataset.head(1)                       
```

#Data Transformation

```python
#renaming variable
dataset = dataset.rename(columns = {'cnt' : 'y'})
dataset.head(1)
```

```python
#index
dataset = dataset.asfreq("D")
dataset.index
```

#Visualization

```python
#viz
dataset["y"].plot(figsize = (10, 7), legend = True)
```

#seasonal decomposition

```python
#library
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#decomposition
decomposition = seasonal_decompose(x = dataset.y,
                                   model = 'multiplicative'
                                  )
decomposition.plot();
```

```python
#seasonality graphs
from statsmodels.graphics.tsaplots import month_plot, quarter_plot
month_plot(dataset.y.resample(rule = 'MS').mean())
quarter_plot(dataset.y.resample(rule = 'Q').mean());
```
